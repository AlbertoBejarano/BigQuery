```python
QUERY = """
SELECT primary_type, year, COUNT(unique_key) AS Crime_Count
FROM  `bigquery-public-data.austin_crime.crime`
GROUP BY primary_type, year
ORDER BY year DESC, Crime_Count DESC
        """
```

```python
Query_Results = bigquery_client.query(QUERY)
df = Query_Results.to_dataframe()
df.head(10)
```

```python
QUERY = """
SELECT source_year AS year, COUNT(is_male) AS birth_count
FROM `bigquery-public-data.samples.natality`
GROUP BY year
ORDER BY year DESC
LIMIT 15""" 
```

```python
Query_Results = bigquery_client.query(QUERY)
df = Query_Results.to_dataframe()
df.head(10)
```

```python
QUERY = """
SELECT country_code, country_name, COUNT(DISTINCT region_code) AS num_regions
FROM  `bigquery-public-data.google_trends.international_top_terms`
WHERE  refresh_date = DATE_SUB(CURRENT_DATE, INTERVAL 1 DAY)
GROUP BY  country_code,  country_name
ORDER BY  num_regions DESC;
"""
```

```python
Query_Results = bigquery_client.query(QUERY)
df = Query_Results.to_dataframe()
df.head(10)
```
